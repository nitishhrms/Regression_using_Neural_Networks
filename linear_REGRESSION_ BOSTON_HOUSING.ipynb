```python

from keras.datasets import boston_housing
from keras.layers.advanced_activations import LeakyReLU
```

```python
Train, Test = boston_housing.load_data()
```

```python
XT,YT = Train
print(XT.shape,YT.shape)
Xt, Yt = Test
print(Xt.shape,Yt.shape)
```

```python
mean = XT.mean(axis=0)
std = XT.std(axis=0)

# Data Normalisation Step
XT = (XT-mean)/std
Xt = (Xt-mean)/std
```

```python
# Build Our Neural Network Model
from keras.models import Sequential
from keras.layers import Dense
import numpy as np
from keras import backend as K

def r2_score(y_true,y_pred):
    
    
    num = K.sum(K.square(y_true-y_pred))
    denom = K.sum(K.square(y_true- K.mean(y_true)))
    return (1-num/denom)

def getModel():
    model = Sequential()
    model.add(Dense(64,activation='relu',input_shape=(13,)))
    model.add(Dense(32,activation='relu'))
    model.add(Dense(1))
    
    model.compile(optimizer='adam',loss='mse',metrics=[r2_score])
    return model
```

```python
model = getModel()
model.summary()
#hist = model.fit(XT,YT,verbose=1,epochs=)
```

```python
hist = model.fit(XT,YT,epochs=400,validation_data=(Xt,Yt),verbose=1)                                                                                                
```

```python
val_loss = hist.history['val_loss']
train_loss = hist.history['loss']
```

```python
import matplotlib.pyplot as plt

plt.plot(val_loss,label="Val Loss ")
plt.plot(train_loss,label="Train Loss")
plt.legend()
plt.show()
```
