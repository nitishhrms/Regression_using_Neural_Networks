```python
def r2_score(y_true,y_pred):
    #y_true=k.array(y_true)
    #y_pred=k.array(y_pred)
    num=k.sum((y_true-y_pred)**2)
    denom=k.sum((y_true-k.mean(y_true))**2)
    return 1-num/denom
```
